# Comment on the datatype of variables

```python
#recast some variables to their correct types.
data['col2']=data['col2'].astype('float')

data['col14']=data['col14'].astype('float')
```

Here both col2 and col14 are wrongly identified as object/string when in reality they belong to float

# The col16 has + and -, replace them ‘P’ and ‘N’ respectively

```python
#encode col16 values from + and - to 'P'and 'N'
data['col16']=data['col16'].map({'+':'P','-':'N'})
```

# Find and display the number of variables of type ‘Object’

```python
#find categorical variables
cat_columns=[c for c in data.columns if data[c].dtypes=='O']
```

```python
data[cat_columns].head()
```

```python
cat_count=len(cat_columns)
print('number of columns of object type:',cat_count)
```

```python
#find numerical variables
num_columns=[c for c in data.columns if data[c].dtypes!='O']
data[num_columns].head()
```

```python
num_count=len(num_columns)
print('number of columns having numerical value',num_count)
```
